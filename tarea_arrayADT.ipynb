```python
class Array:
    def __init__( self, tam ):
        self.__tamanio = tam
        self.__data = [ 0 for x in range(self.__tamanio)] #Generador: Estuctura especial de Phyton que ayuda a crear listas de manera automática 

    def to_string( self ):
        info = ""
        for dato in self.__data:
            info = info + str(dato) + " | "
        return info

    def get_item( self, x, y ):
        return self.__data[x][y]

    def get_lenght( self ):
        return len(self.__data)

    def set_item( self, dato, index ):
        #Hace falta revisar los límites del índice 
        if index >= 0 and index < self.get_lenght():
            self.__data[index] = dato
        else:
            print("Error: índice fuera de rango")

    def clear( self, dato ):
         self.__data = [ dato for x in range(self.__tamanio)]
```

```python
class Trabajador:
    def __init__(self , nt, nom, pat, mat, he, sb, ai):
        self.__numero_trabajador = int(nt) 
        self.__nombres = nom
        self.__apellido_paterno = pat
        self.__apellido_materno = mat
        self.__horas_extras = int(he)
        self.__sueldo_base = int(sb)
        self.__anio_ingreso = int(ai)
        #asi para todos los atributos 

    # aqui van todos los métodos de acceso
    def set_numero_trabajador( self, nt):
        self.__numero_trabajador = nt 

    def get_numero_trabajador( self ):
        print(f"El numero de trabajador es: {self.__numero_trabajador}")
        
    def set_nombres( self, nom):
        self.__nombres = nom 

    def get_nombres( self ):
        print(f"El nombre del trabajador es: {self.__nombres}")
        
    def set_apellido_paterno( self, pat):
        self.__apellido_paterno = pat 

    def get_apellido_paterno( self ):
        print(f"El apellido paterno del trabajador es: {self.__apellido_paterno}")
        
    def set_apellido_materno( self, mat):
        self.__apellido_materno = mat 

    def get_apellido_materno( self ):
        print(f"El apellido materno del trabajador es: {self.__apellido_materno}")
        
    def set_horas_extras( self, he):
        self.__horas_extras = he 

    def get_horas_extras( self ):
        print(f"Horas extras trabajadas: {self.__horas_extras}")

    def set_sueldo_base( self, sb):
        self.__sueldo_base = sb 

    def get_sueldo_base( self ):
        print(f"El sueldo base del trabajador es: ${self.__sueldo_base}")
        
    def set_anio_ingreso( self, ai):
        self.__anio_ingreso = ai

    def get_anio_ingreso( self ):
        print(f"Labora desde el año: {self.__anio_ingreso}")
    
    def to_string( self ):
        print(f"Num. Trabajador: {self.__numero_trabajador}\nNombre: {self.__nombres}\nApellidos: {self.__apellido_paterno} {self.__apellido_materno}\nHoras extra trabajadas: {self.__horas_extras}\nSueldo base: ${self.__sueldo_base}\nLaborando desde: {self.__anio_ingreso}")
               
    # métodos de uso gral --> metodos que hacen cálculos: horas extra, etc.

    def calcular_antiguedad( self ):
        antiguedad = 2021 - self.__anio_ingreso
        print(f"---> ANTIGUEDAD: {antiguedad} AÑOS")
    
    def sueldo_con_prestacion( self ):
        antiguedad = 2021 - self.__anio_ingreso
        prestacion = self.__sueldo_base * (antiguedad * 0.03)
        sueldo_nuevo = self.__sueldo_base + prestacion
        print(f"---> SUELDO CON PRESTACIÓN: ${sueldo_nuevo}")

    def pago_horas_extras( self ):
        pago_horas_extra = self.__horas_extras * 276.5
        print(f"---> PAGO POR {self.__horas_extras} HORA(S) EXTRA: ${pago_horas_extra}")

for i in range(1, len(info)):
    print(f"---------- [TRABAJADOR {i}] ----------")
    trab = Trabajador(info[i][0], info[i][1], info[i][2], info[i][3], info[i][4], info[i][5], info[i][6])
    trab.to_string()
    trab.calcular_antiguedad()
    trab.sueldo_con_prestacion()
    trab.pago_horas_extras()
    print("\n")

```

```python
class NominaADT:
    def __init__( self, ruta_archivo ):
        info = []
        self.__num_empleados = 0
        archivo = open(ruta_archivo, 'rt', encoding='latin-1' ) 
        
        for i in range(15):
            lista = archivo.readline().split(',')
            for index in range(len(lista)): #Quita espacios y saltos de linea
                lista[index] = lista[index].strip()
            info.append(lista)
        info.pop(0)

        self.__num_empleados = int(len(info))
        self.__data = Array(self.__num_empleados)
        print(f"Numero de empleados: {self.__num_empleados}")

        #Ingresamos los datos a self.__data con los métodos de ArrayADT
        for i in range(self.__num_empleados):
            self.__data.set_item(info[i], i)
            
    def calcular_nomina( self ):
        sueldos_con_prestacion = []
        pago_horas_extra = []
        for i in range (self.__num_empleados):
            antiguedad = 2021 - int( self.__data.get_item( i, 6) )
            prestacion = int( self.__data.get_item( i, 5) ) * (antiguedad * 0.03)
            sueldo_actual = int( self.__data.get_item( i, 5 ) ) + prestacion
            sueldos_con_prestacion.append(sueldo_actual)
        
        for i in range (self.__num_empleados):
            pago_he = antiguedad = 276.5 * int(self.__data.get_item( i, 4 ))
            pago_horas_extra.append(pago_he)
        
        total_sueldos = sum(sueldos_con_prestacion)
        total_he = sum(pago_horas_extra)

        print("\n----- PAGO DE NÓMIMA DEL MES -----")
        for i in range (self.__num_empleados):
            print(f"No. Empleado: {self.__data.get_item( i, 0 ) } | Sueldo actual (considerando prestación): ${sueldos_con_prestacion[i]} | Pago por {self.__data.get_item( i, 4 )} horas extra: ${pago_horas_extra[i]}")

        print(f"\t\t\t\t\t\t      Total: ${total_sueldos} |\t\t Total: ${total_he}")
        print(f"------> TOTAL PAGO DE NÓMINA: ${total_sueldos + total_he}")    

    def buscar_mayor_antiguedad( self ):
        anios_ingreso = []
        for i in range (self.__num_empleados):
            anios_ingreso.append( int(self.__data.get_item(i,6)))
    
        mas_antiguo = min(anios_ingreso)
        indices = []
        for i in range( len(anios_ingreso) ):
            if anios_ingreso[i] == mas_antiguo:
                indices.append(i)
        
        print("\n----- Empleado(s) con MAYOR antiguedad ----- \n")
        for i in indices:
            print(f"No. de Empleado: {self.__data.get_item( i, 0 ) } \nNombre Completo: {self.__data.get_item(i, 1) } {self.__data.get_item( i, 2) } {self.__data.get_item( i, 3) }")
            print(f"Horas extra: {self.__data.get_item( i, 4) } \nSueldo Base: {self.__data.get_item( i, 5 )}\nAño de ingreso: {self.__data.get_item(i, 6)}")
            print("---------------------")
  
    def buscar_menor_antiguedad( self ): 
        anios_ingreso = []
        for i in range (self.__num_empleados):
            anios_ingreso.append( int(self.__data.get_item(i,6)))
    
        mas_antiguo = max(anios_ingreso)
        indices = []
        for i in range( len(anios_ingreso) ):
            if anios_ingreso[i] == mas_antiguo:
                indices.append(i)
        
        print("\n----- Empleado(s) con MENOR antiguedad ----- \n")
        for i in indices:
            print(f"No. de Empleado: {self.__data.get_item( i, 0 ) } \nNombre Completo: {self.__data.get_item(i, 1) } {self.__data.get_item( i, 2) } {self.__data.get_item( i, 3) }")
            print(f"Horas extra: {self.__data.get_item( i, 4) } \nSueldo Base: {self.__data.get_item( i, 5 )}\nAño de ingreso: {self.__data.get_item(i, 6)}")
            print("---------------------")


nomi = NominaADT('junio.dat')
nomi.calcular_nomina()
nomi.buscar_mayor_antiguedad()
nomi.buscar_menor_antiguedad()
```
